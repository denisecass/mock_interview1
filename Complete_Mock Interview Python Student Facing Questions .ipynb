<b> Q4. Please create two bar plots as per below that show:
    
    1) The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
    2) The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States 

Please note you will be looking specifically at the *Job Role* column

<b> Put Your Code Below </b>

```python
df["Degree Status"].value_counts()
```

```python
HS_grad = df[(df["Degree Status"]==" HS-grad")]
HS_grad
```

```python
# 1st Graph: The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
grad_earnings = HS_grad[HS_grad["Earnings"]==" <=50K"]
cnt = grad_earnings["Job-Role"].value_counts()
x = list(cnt.index)
y = list(cnt.values)
print(x,y)

plt.bar(x,y,width=0.5,)
plt.xticks(rotation=90)
plt.title("Individuals who earn <50k, have a High School Graduate Diploma and are in the United States")
plt.show()
```

```python
# 2nd Graph: The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States 
grad_earnings = HS_grad[HS_grad["Earnings"]==" >50K"]
cnt = grad_earnings["Job-Role"].value_counts()
x = list(cnt.index)
y = list(cnt.values)
print(x,y)

plt.bar(x,y,width=0.5,)
plt.xticks(rotation=90)
plt.title("Individuals who earn >50k, have a High School Graduate Diploma and are in the United States")
plt.show()
```

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
highest_earning_job_role = df['Job-Role'][df['Earnings'] == ' >50K'].value_counts()
highest_earning_job_role
```

```python
x = list(highest_earning_job_role.index)
y = list(highest_earning_job_role.values)
print(x,y)

plt.bar(x,y,width=0.5,)
plt.xticks(rotation=90)
plt.title("Highest Earning Job Roles")
plt.show()
```

```python
import numpy as np
y = np.array(highest_earning_job_role)
mylabels = ["Exec-managerial","Prof-specialty", "Sales","Craft-repair","Adm-clerical","Transport-moving","Tech-support","Machine-op-inspct","Protective-serv","?","Other-service","Farming-fishing","Handlers-cleaners","Armed-Forces","Priv-house-serv"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Highest Earning Job Roles:")
plt.axis('equal')
plt.show() 
```
